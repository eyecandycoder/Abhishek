```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X= dataset.iloc[: , : -1].values
y = dataset.iloc[: , -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[30, 87000]])))
```

```python
y_pred = classifier.predict(X_test)
np.set_printoptions(precision =2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1),)
```

**Confusion matrix and accuracy**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

**Visualization of test set and prediction**


```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train),y_train
X1, X2 = np.meshgrid( np.arange( start = X_set[ : , 0].min() -10 , stop = X_set[ : , 0].max() +10, step = 0.25 ) ,
                     np.arange( start = X_set[ : , 1].min() -1000 , stop = X_set[ : , 1].max() +1000, step = 0.25 ))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape), alpha = 0.75,
                                        cmap = ListedColormap(('red' , 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set == j,0] , X_set[y_set == j,1], c = ListedColormap(('red' , 'green'))(i), label = j)
plt.title(' Logistic Regression(Train_Dataset)')
plt.xlabel('Age')
plt.ylabel('Estimated_Salary')
plt.legend()
plt.show()
```
